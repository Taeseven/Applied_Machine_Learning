```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load data
data = np.loadtxt('iris.data', delimiter = ',', usecols = range(4))
label = np.hstack((['r']*50, ['g']*50, ['b']*50))
```

```python
% matplotlib inline
# plot figure
color_map = {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica':'b'}

plt.figure(figsize=(10,10))
plt.suptitle('Iris Data (red=setosa, green=versicolor, blue=virginica)', fontsize=15, fontweight='bold')
# first row
plt.subplot(4, 4, 1)
plt.annotate('Sepal.Length', xy=(0.25, 0.5))
plt.subplot(4, 4, 2)
plt.scatter(data[:,1], data[:,0], c=label ,marker='.')
plt.subplot(4, 4, 3)
plt.scatter(data[:,2], data[:,0], c=label ,marker='.')
plt.subplot(4, 4, 4)
plt.scatter(data[:,3], data[:,0], c=label ,marker='.')

# second row
plt.subplot(4, 4, 5)
plt.scatter(data[:,0], data[:,1], c=label ,marker='.')
plt.subplot(4, 4, 6)
plt.annotate('Sepal.Width', xy=(0.25, 0.5))
plt.subplot(4, 4, 7)
plt.scatter(data[:,2], data[:,1], c=label ,marker='.')
plt.subplot(4, 4, 8)
plt.scatter(data[:,3], data[:,1], c=label ,marker='.')

# thrid row
plt.subplot(4, 4, 9)
plt.scatter(data[:,0], data[:,2], c=label ,marker='.')
plt.subplot(4, 4, 10)
plt.scatter(data[:,1], data[:,2], c=label ,marker='.')
plt.subplot(4, 4, 11)
plt.annotate('Petal.Legth', xy=(0.25, 0.5))
plt.subplot(4, 4, 12)
plt.scatter(data[:,3], data[:,2], c=label ,marker='.')

# fourth row
plt.subplot(4, 4, 13)
plt.scatter(data[:,0], data[:,3], c=label ,marker='.')
plt.subplot(4, 4, 14)
plt.scatter(data[:,1], data[:,3], c=label ,marker='.')
plt.subplot(4, 4, 15)
plt.scatter(data[:,2], data[:,3], c=label ,marker='.')
plt.subplot(4, 4, 16)
plt.annotate('Petal.Width', xy=(0.25, 0.5))
plt.savefig("iris.png")
```
